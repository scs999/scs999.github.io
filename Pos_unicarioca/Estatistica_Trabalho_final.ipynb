**Amostras da população**

    > Tamanhos: 100, 1000 e 4000
    > Utilizando função

```python
def gerar_amostras(populacao, tamanhos_amostra):
    amostras_por_tamanho = {}

    for tamanho_amostra in tamanhos_amostra:

        idx_amostras = np.random.choice(populacao.index, size=tamanho_amostra, replace=False)
        amostras = populacao.loc[idx_amostras]
        amostras_por_tamanho[tamanho_amostra] = amostras

    return amostras_por_tamanho

populacao = pd.DataFrame(df)
populacao.set_index('MUNICÍPIO', inplace=False)

tamanhos_amostra = [100,1000,4000]  # tamanho das amostras
amostras_por_tamanho = gerar_amostras(populacao, tamanhos_amostra)

# Exibindo dados das amostras
for tamanho, amostras in amostras_por_tamanho.items():
    print(f"\nTamanho da amostra: {tamanho}:")
    print("-----------------------------------------------------------------")
    print(amostras)
    print("\n-------------------- ESTIMADORES ------------------------")
    pop_amt = amostras['POP']
    pop_total_amt = pop_amt.sum()
    print(f"POPULAÇÃO TOTAL: {pop_total_amt}")
    pop_media_amt = pop_amt.mean()
    print("MÉDIA DA POPULAÇÃO:  {:.0f}".format(pop_media_amt))
    pop_mediana_amt =  pop_amt.median()
    print(f"MEDIANA DA POPULAÇÃO: {pop_mediana_amt}")
    # Menor população
    min_index_amt = pop_amt.idxmin()
    menor_mun_nome_amt = amostras.loc[min_index_amt, 'MUNICÍPIO']
    menor_mun_pop_amt = amostras.loc[min_index_amt, 'POP']
    print(f"Município com menor população: {menor_mun_nome_amt} com {menor_mun_pop_amt} habitantes.")
    #Maior população
    max_index_amt = pop_amt.idxmax()
    maior_mun_nome_amt = amostras.loc[max_index_amt, 'MUNICÍPIO']
    maior_mun_pop_amt = amostras.loc[max_index_amt, 'POP']
    print(f"Município com maior população: {maior_mun_nome_amt} com {maior_mun_pop_amt} habitantes.")
    print("\n-----------------------------------------------------------------")
    varp  = st.pvariance(pop_amt)
    dpp = st.pstdev(pop_amt)
    print("Variância: {:.0f}".format(varp))
    print("Desvio Padrão :{:.0f}".format(dpp))
    cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100
    print("Coeficiente de variação: {:.4f}".format(cv(pop_amt)))
    print("\n-----------------------------------------------------------------")
```

**Fração amostral**

    >  Tamanhos: 10%, 20% e 100%
    >  Utilizando função



```python
def obter_amostras_por_fracao(populacao, fracoes):
    amostras_por_fracao = {}

    for fracao in fracoes:
        tamanho_amostra = int(fracao * len(populacao))
        indices_amostras = np.random.choice(populacao.index, size=tamanho_amostra, replace=False)
        amostras_fr = populacao.loc[indices_amostras]
        amostras_por_fracao[fracao] = amostras_fr

    return amostras_por_fracao

populacao = pd.DataFrame(df)
populacao.set_index('MUNICÍPIO', inplace=False)

fracoes_amostra = [0.1, 0.2, 1]
amostras_por_fracao = obter_amostras_por_fracao(populacao, fracoes_amostra)

# Exibir as amostras
for fracao, amostras_fr in amostras_por_fracao.items():
    print(f"\nAmostras com {int(fracao * 100)}% da população:")
    print("-----------------------------------------------------------------")
    print(amostras_fr)
    pop_amt_fr = amostras_fr['POP']
    print("\n-------------------- ESTIMADORES ------------------------")
    pop_total_amt = pop_amt_fr.sum()
    print("POPULAÇÃO TOTAL:" , pop_total_amt)
    pop_media_amt = pop_amt_fr.mean()
    print("MÉDIA DA POPULAÇÃO:  {:.0f}".format(pop_media_amt))
    pop_mediana_amt = pop_amt_fr.median()
    print("MEDIANA DA POPULAÇÃO: ", pop_mediana_amt)
    min_index_amt = pop_amt_fr.idxmin()
    menor_mun_nome_amt = amostras_fr.loc[min_index_amt, 'MUNICÍPIO']
    menor_mun_pop_amt = amostras_fr.loc[min_index_amt, 'POP']
    print(f"Município com menor população: {menor_mun_nome_amt} com {menor_mun_pop_amt} habitantes.")
    max_index_amt = pop_amt_fr.idxmax()
    maior_mun_nome_amt = amostras_fr.loc[max_index_amt, 'MUNICÍPIO']
    maior_mun_pop_amt = amostras_fr.loc[max_index_amt, 'POP']
    print(f"Município com maior população: {maior_mun_nome_amt} com {maior_mun_pop_amt} habitantes.")
    varp  = st.pvariance(pop_amt_fr)
    dpp = st.pstdev(pop_amt_fr)
    print("Variância: {:.0f}".format(varp))
    print("Desvio Padrão: {:.0f}".format(dpp))
    cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100
    print("Coeficiente de variação: {:.4f}".format(cv(pop_amt_fr)))
    print("\n-----------------------------------------------------------------")
```

# 4 - Intervalo de confiança para cada amostra
        Determine o Intervalo de Confiança de 95% para a Média (habitantes/município)


    > Tamanhos: 100, 1000 e 4000
    > Utilizando função

```python
for tamanho, amostras in amostras_por_tamanho.items():
    print(f"\nTamanho da amostra: {tamanho}:")
# Media da amostra e Intervalo de confiança
    media_amostra = np.mean(amostras['POP'])
    desvio_padrao_amostral = np.std(amostras['POP'], ddof=1)
    erro_padrao = desvio_padrao_amostral / np.sqrt(len(amostras['POP']))
    z_valor = stats.norm.ppf(0.975)
    margem_erro = z_valor * erro_padrao
    intervalo_confianca = (media_amostra - margem_erro, media_amostra + margem_erro)
    print(f"Média da Amostra:{media_amostra}")
    print(f"Intervalo de Confiança (95%): {intervalo_confianca}")
    print("Margem:",margem_erro)
    print("\n -----------------------------------------------------------------")
```
